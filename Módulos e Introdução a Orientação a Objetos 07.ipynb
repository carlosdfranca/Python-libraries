# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
```

```python
plt.plot(meses, vendas, 'bo--')
plt.axis([0, 50, 0, max(vendas)+200])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas, color='yellow')
plt.show()
```

- Barras

```python
plt.bar(meses, vendas, color = 'red')
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(figsize=(15, 3))

plt.subplot(131)
plt.plot(meses, vendas, 'ro--')
plt.subplot(132)
plt.bar(meses, vendas)
plt.subplot(133)
plt.scatter(meses,vendas,color='black')
plt.suptitle('Gráficos de Vendas', fontsize = '18')
plt.show()
```
